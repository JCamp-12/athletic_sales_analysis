```python
# Import Libraries and Dependencies
import pandas as pd
from pathlib import Path
```

### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.
sales_2020_path = Path('Resources/athletic_sales_2020.csv')
sales_2021_path = Path('Resources/athletic_sales_2021.csv')

# Read in data and index by date
sales_2020_data = pd.read_csv(sales_2020_path, index_col='invoice_date')
sales_2021_data = pd.read_csv(sales_2021_path, index_col='invoice_date')
```

```python
# Display the 2020 sales DataFrame
sales_2020_data
```

```python
# Display the 2021 sales DataFrame
sales_2021_data
```

#### Check the data types of each DataFrame

```python
# Check the 2020 sales data types.
sales_2020_data.dtypes
```

```python
# Check the 2021 sales data types.
sales_2021_data.dtypes
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
joined_data_rows = pd.concat([sales_2020_data, sales_2021_data], axis='rows', join='inner')
joined_data_rows = joined_data_rows.reset_index(drop=False)
joined_data_rows
```

```python
# Check if any values are null.
null_presence = joined_data_rows.isnull().any()
null_presence
```

```python
# Check the data type of each column
joined_data_rows.dtypes
```

```python
# Convert the "invoice_date" to a datetime datatype
joined_data_rows['invoice_date'] = pd.to_datetime(joined_data_rows['invoice_date'], format='%m/%d/%y')
joined_data_rows
```

```python
# Confirm that the "invoice_date" data type has been changed.
joined_data_rows.dtypes
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
# Rename the sum to "Total_Products_Sold".
# The object returned is a "GroupBy" object and cannot be viewed normally...
grouped_data_rows = joined_data_rows.groupby(['region', 'state', 'city'])[["units_sold"]].agg('sum')
grouped_data_rows.rename(columns={'units_sold': 'Total_Products_Sold'}, inplace=True)

# Show the top 5 results.
grouped_data_rows.sort_values(by='Total_Products_Sold', ascending=False).head(5)         
```

#### Using `pivot_table`

```python
# Show the number products sold for region, state, and city.
#  Pivot on the date_ending with the book_name as the index, and pass the "total_sales" as the values.
pt_joined_data_rows = pd.pivot_table(joined_data_rows, 
                                   index=['region','state', 'city'],
                                   values='units_sold',
                                   aggfunc='sum')
# Show the table.
pt_joined_data_rows

# Rename the "units_sold" column to "Total_Products_Sold"
pt_joined_data_rows = pt_joined_data_rows.rename(columns={"units_sold": "Total_Products_Sold"})

# Show the top 5 results.
pt_joined_data_rows.sort_values(by='Total_Products_Sold', ascending=False).head()
```

### 3. Determine which Region had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each region, state, and city.
# Rename the "total_sales" column to "Total Sales"
grouped_data_rows_total_sales = joined_data_rows.groupby(['region', 'state', 'city'])[["total_sales"]].agg('sum')
grouped_data_rows_total_sales.rename(columns={'total_sales': 'Total Sales'}, inplace=True)

# Show the top 5 results.
grouped_data_rows_total_sales.sort_values(by='Total Sales', ascending=False).head(5) 
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each region, state, and city.

pt_joined_data_rows_total_sales = pd.pivot_table(joined_data_rows, 
                                   index=['region','state', 'city'],
                                   values='total_sales',
                                   aggfunc='sum')


# Optional: Rename the "total_sales" column to "Total Sales"
pt_joined_data_rows_total_sales = pt_joined_data_rows_total_sales.rename(columns={"total_sales": "Total Sales"})

# Show the top 5 results.
pt_joined_data_rows_total_sales.sort_values(by='Total Sales', ascending=False).head()
```

### 4. Determine which Retailer had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each retailer, region, state, and city.
# Rename the "total_sales" column to "Total Sales"
grouped_data_rows_retailer = joined_data_rows.groupby(['retailer', 'region', 'state', 'city'])[["total_sales"]].agg('sum')
grouped_data_rows_retailer.rename(columns={'total_sales': 'Total Sales'}, inplace=True)

# Show the top 5 results.
grouped_data_rows_retailer.sort_values(by='Total Sales', ascending=False).head(5)  

```

```python
# Show the total sales for the products sold for each retailer, region, state, and city.

pt_joined_data_rows_total_sales_retailer = pd.pivot_table(joined_data_rows, 
                                   index=['retailer', 'region','state', 'city'],
                                   values='total_sales',
                                   aggfunc='sum')


# Optional: Rename the "total_sales" column to "Total Sales"
pt_joined_data_rows_total_sales_retailer = pt_joined_data_rows_total_sales_retailer.rename(columns={"total_sales": "Total Sales"})

# Show the top 5 results.
pt_joined_data_rows_total_sales_retailer.sort_values(by='Total Sales', ascending=False).head()
```

### 5. Determine which Retailer Sold the Most Women's Athletic Footwear

```python
# Filter the sales data to get the women's athletic footwear sales data.
womens_joined_data_rows = joined_data_rows[joined_data_rows['product'] == 'Women\'s Athletic Footwear']
womens_joined_data_rows
```

#### Using `groupby`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
grouped_data_rows_womens_ath_foo = womens_joined_data_rows.groupby(['retailer', 'region', 'state', 'city'])[["units_sold"]].agg('sum')
grouped_data_rows_womens_ath_foo.rename(columns={'units_sold': 'Womens_Footwear_Units_Sold'}, inplace=True)

# Show the top 5 results.
grouped_data_rows_womens_ath_foo.sort_values(by='Womens_Footwear_Units_Sold', ascending=False).head(5) 
```

#### Using `pivot_table`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.

pt_joined_data_rows_womens_ath_foo = pd.pivot_table(womens_joined_data_rows, 
                                   index=['retailer', 'region','state', 'city'],
                                   values='units_sold',
                                   aggfunc='sum')

# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
pt_joined_data_rows_womens_ath_foo = pt_joined_data_rows_womens_ath_foo.rename(columns={"units_sold": "Womens_Footwear_Units_Sold"})

# Show the top 5 results.
pt_joined_data_rows_womens_ath_foo.sort_values(by='Womens_Footwear_Units_Sold', ascending=False).head()
```

### 5. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.

pt_joined_data_rows_womens_ath_foo_day = pd.pivot_table(womens_joined_data_rows, 
                                   index=['invoice_date'],
                                   values='total_sales',
                                   aggfunc='sum')

# Optional: Rename the "total_sales" column to "Total Sales"
pt_joined_data_rows_womens_ath_foo_day = pt_joined_data_rows_womens_ath_foo_day.rename(columns={"total_sales": "Total Sales"})

# Show the table.
pt_joined_data_rows_womens_ath_foo_day.sort_values(by='Total Sales', ascending=False).head(10)
```

```python
# Resample the pivot table into daily bins, and get the total sales for each day.

# womens_joined_data_rows.dtypes
total_sales_daily_resample = pt_joined_data_rows_womens_ath_foo_day.resample('D').sum()


# Sort the resampled pivot table in ascending order on "Total Sales".
total_sales_daily_resample.sort_values(by='Total Sales', ascending=False).head(10)
```

### 6.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
# Resample the pivot table into weekly bins, and get the total sales for each week.
total_sales_weekly_resample = pt_joined_data_rows_womens_ath_foo_day.resample('W').sum()

# Sort the resampled pivot table in ascending order on "Total Sales".
total_sales_weekly_resample.sort_values(by='Total Sales', ascending=False).head(10)
```
